```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import calinski_harabasz_score
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=500, centers=3, cluster_std=0.60, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("Clusters encontrados")
plt.show()
```

```python
sil_score = silhouette_score(X, labels)
print("Silhouette Score:", sil_score)
```

```python
db_score = davies_bouldin_score(X, labels)
print("Davies-Bouldin:", db_score)
```

```python
ch_score = calinski_harabasz_score(X, labels)
print("Calinski-Harabasz:", ch_score)
```

```python
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)

    print(f"k = {k}")
    print("  Silhouette:", silhouette_score(X, labels))
    print("  Davies-Bouldin:", davies_bouldin_score(X, labels))
    print("  Calinski-Harabasz:", calinski_harabasz_score(X, labels))
    print()
```
